```python
import pandas as pd
```

```python
title_df = pd.read_csv('title.tsv', sep='\t')
episode_df = pd.read_csv('episode.tsv', sep='\t')
ratings_df = pd.read_csv('ratings.tsv', sep='\t')
```

```python
title_df.head()
```

```python
episode_df.head()
```

```python
ratings_df.head()
```

# Squid Game

```python
squid_game_title_df = title_df[title_df['primaryTitle'] == 'Squid Game']
squid_game_title_df = squid_game_title_df.reset_index()
squid_game_title_df = squid_game_title_df[['tconst','primaryTitle']]
squid_game_title_df
```

```python
squid_game_title_df = squid_game_title_df.rename(columns={'tconst': 'parentTconst'})
squid_game_title_df
```

```python
new_squidgame_df = pd.merge(squid_game_title_df,episode_df,on='parentTconst')
new_squidgame_df
```

```python
new_squidgame_df = pd.merge(new_squidgame_df,ratings_df,on='tconst')
new_squidgame_df
```

```python
new_squidgame_df = new_squidgame_df.sort_values(by="episodeNumber", ascending=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.ylim(0, 10)
ax = sns.lineplot(data=new_squidgame_df, x="episodeNumber", y="averageRating")
ax.set(xlabel='Episode', ylabel='Average Rating')
```

# Alice In Borderland

```python
aliceb_title_df = title_df[title_df['primaryTitle'] == 'Alice in Borderland']
aliceb_title_df = aliceb_title_df.reset_index()
aliceb_title_df = aliceb_title_df[['tconst','primaryTitle']]
aliceb_title_df
```

```python
aliceb_title_df = aliceb_title_df.rename(columns={'tconst': 'parentTconst'})
aliceb_title_df
```

```python
new_aib_df = pd.merge(aliceb_title_df,episode_df,on='parentTconst')
new_aib_df
```

```python
new_aib_df = pd.merge(new_aib_df,ratings_df,on='tconst')
new_aib_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.ylim(0, 10)
ax = sns.lineplot(data=new_aib_df, x="episodeNumber", y="averageRating")
ax.set(xlabel='Episode', ylabel='Average Rating')
```

# Combine

```python
result = new_squidgame_df.append(new_aib_df)
result
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,12))
plt.ylim(0, 10)
ax = sns.lineplot(data=result, x="episodeNumber", y="averageRating", hue="primaryTitle", marker="o")
ax.set_title("Squid Game vs Alice In Borderland in terms of Episode Ratings", fontsize=15)
ax.set(xlabel='Episode', ylabel='Average Rating')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
ax.text(8.5, 1,'Source: iMDB', fontsize=9)
ax.text(8.5, 0.8,'Created by: Kok Jim Meng', fontsize=9)
```
